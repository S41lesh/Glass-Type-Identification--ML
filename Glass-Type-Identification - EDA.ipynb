# **Train Test Split**

````python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.25, random_state=42)
````

````python
y_train = y_train.values.ravel()
y_test = y_test.values.ravel()
````

### **Shape of Train and Test Set**

````python
print('Shape of X_train = ' + str(X_train.shape))
print('Shape of X_test = ' + str(X_test.shape))
print('Shape of y_train = ' + str(y_train.shape))
print('Shape of y_test = ' + str(y_test.shape))
````

# **Modelling Dataset**

### **Non-Linear SVM**

````python
from sklearn.svm import SVC
import time

start = time.time()
svm_model = SVC(gamma='auto').fit(X_train, y_train)
svm_predictions = svm_model.predict(X_test)

acc_train = svm_model.score(X_train, y_train)
acc_test = svm_model.score(X_test, y_test)
end = time.time()
print("Accuracy for training set: ",acc_train)
print("Accuracy for test set: ",acc_test)
print((end - start)*10**3)



````

### **Desicion Tree**

````python
from sklearn.tree import DecisionTreeClassifier
import time
start = time.time()
tree_model = DecisionTreeClassifier(random_state=0).fit(X_train, y_train)
tree_predictions = tree_model.predict(X_test)
end = time.time()
print("Accuracy for training set: ",tree_model.score(X_train,y_train))
print("Accuracy for test set: ",tree_model.score(X_test,y_test))
print((end - start)*10**3)
````

### **Plot for Decision Tree**

````python
from sklearn import tree

fig, axes = plt.subplots(nrows = 1,ncols = 1,figsize = (100,100), dpi=300)
tree.plot_tree(tree_model)
````

**K Nearest Neighbour**

````python
from sklearn.metrics import accuracy_score
from sklearn.neighbors import KNeighborsClassifier
import time
start = time.time()
Scores = []

for i in range (2,11):
    knn = KNeighborsClassifier(n_neighbors=i)
    knn.fit(X_train, y_train)
    score = knn.score(X_test,y_test)
    Scores.append(score)
end = time.time()
print(knn.score(X_train,y_train))

print(knn.score(X_test,y_test))
print((end - start)*10**3)
````

**Random Forest**

````python
from sklearn.ensemble import RandomForestClassifier
import time
start = time.time()
Scores = []

forest = RandomForestClassifier(criterion='gini', n_estimators=10, min_samples_leaf=5, min_samples_split=5,  random_state=1, n_jobs=-1)

forest.fit(X_train, y_train)
end = time.time()
score = forest.score(X_test,y_test)

print(forest.score(X_train,y_train))
print(score)
print((end - start)*10**3)
````
